Por experiencia sabemos que no queremos en nuestro modelo, varibales con alta cardinalidad porque adem√°s de hacer m√°s complejo el proceso, tambi√©n puede inducir overfitting. Tampoco queremos variables con demasiados valores nulos, pues al imputar tantos valores podemos introducir ruido innecesario a nuestro proceso. Variables con alta correlaci√≥n entre s√≠ pueden no aportar mucha informaci√≥n extra, as√≠ que podemos dejar solo algunas; tampoco queremos varibales que tengan valores iguales en todos los registros, por ejmplo, si hay variables con valor de 1 en casi todos los datos, es posible que al eliminarla no afecte en la creaci√≥n de nuestros modelos.

Y finalmente debemos tener la capacidad de identificar variables que, por su distribuci√≥n no son malas, pero que no aportan a nuestro objetivo, por ejemplo variables ID, no aportan valor predictivo nuestros modelos, porque solo son identificadoras.

```python
# Primero identificamos las variables categ√≥ricas
categorical_columns = data_total.select_dtypes(include=['object']).columns

# Calculamos la cardinalidad de cada variable categ√≥rica
cardinalities = {col: data_total[col].nunique() for col in categorical_columns}

# las mostramos
for col, cardinality in cardinalities.items():
    print(f"Variable: {col}, Cardinalidad: {cardinality}")

# Analizamos la distribuci√≥n de la cardinalidad de una mejor forma
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
plt.hist(cardinalities.values(), bins=30, edgecolor='k')
plt.xlabel('N√∫mero de valores √∫nicos')
plt.ylabel('Frecuencia')
plt.title('Distribuci√≥n de la cardinalidad de las variables categ√≥ricas')
plt.show()
```

Podemos darnos cuenta de hay contadas variables con alta cardinalidad, veamos m√°s a detalle de cuales se trata.

```python
# Calculamos las variables con alta cardinalidad
high_cardinality_cols = [col for col in categorical_columns if data_total[col].nunique() > 50]

print(f"Variables categ√≥ricas con alta cardinalidad: {high_cardinality_cols}")
```

```python
data_total[high_cardinality_cols].sample(5)
```

Ahora procedemos a buscar variables con muchos valores nulos, no queremos agregar ruido en exceso a nuestros datos al momento de imputar tantos datos.  

```python
# Calculamos el porcentaje de valores nulos por columna
null_percentages = data_total.isnull().mean() * 100

# Mostramos los porcentajes de valores nulos
print(null_percentages)
```

Podemos ver que hay casos donde el 74% de los datos son nulos, incluirlos en lugar de beneficiarnos nos har√≠a agregar valores que pueden no ir acorde a la realidad y nos estar√≠amos alejando de nuestro objetivo, vamos a establecer un umbral del 50% y veamos que columnas tendr√≠amos que eliminar

```python
# Establecemos un umbral (inicialmente del 50%, podemos bajarlo incluso)
threshold = 50

# Identificamos columnas a eliminar
cols_to_drop = null_percentages[null_percentages > threshold].index

print(f"Columnas a eliminar por tener m√°s del {threshold}% de valores nulos: {cols_to_drop}")
```

```python
data_total[cols_to_drop].info()
```

```python
# sorprendentemente hay mas de 200 columnas con m√°s del 50% de datos nulos
data_total[cols_to_drop].sample(5)
```

Sin embargo, un umbral del 50% es comunmente aceptado en este tipo de casos, queremos modelos lo mas simples posibles, con la mayor cantidad de informaci√≥n util, pero sin agregar ruido en exceso, decidimos eliminarlo, si vemos que nuestros modelos no generan buenas predicciones con los datos que dejamos, siempre podemos volver a experimentar el eliminar menos columnas, la Ciencia de Datos es un proceso iterativo y experimental.

```python
# Eliminamos las columnas con exceso de nulos
data_total_cleaned = data_total.drop(columns=cols_to_drop)
```

```python
data_total_cleaned.info()
```

Vamos ahora a buscar variables constantes, que justo por esa cualidad, carecen de valor predictivo.

```python
# Calculamos la desviaci√≥n est√°ndar de cada columna
std_devs = data_total_cleaned.select_dtypes(include='number').std()

# Identificamos columnas con desviaci√≥n est√°ndar cercana a cero (por ejemplo menores a 0.01)
constant_columns = std_devs[std_devs < 0.01].index

print(f"Columnas constantes a eliminar: {constant_columns}")
```

```python
data_total_cleaned[constant_columns].sample(5)
```

```python
# Estas variables parecen ser constantes, no aportan valor predictivo, asi que la eiminamos
data_total_cleaned = data_total_cleaned.drop(columns=constant_columns)
```

Veamos ahora si hay variables muy correlacionadas entre s√≠

```python
# Calculamos la matriz de correlaci√≥n
corr_matrix = data_total_cleaned.select_dtypes(include='number').corr().abs()

import seaborn as sns
# Mostramos la matriz de correlaci√≥n, esperemos que sea posible apreciarse
plt.figure(figsize=(12, 10))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Matriz de Correlaci√≥n')
plt.show()
```

Vemos que hay mucho trabajo por hacer en este sentido, vamos a dejar una columna de cada grupo de columnas que esten altamente correlacionadas entre s√≠.

```python
import numpy as np
# Seleccionamos el umbral de correlaci√≥n (en este caso de 0.8)
threshold = 0.8

# Identificamos columnas que tienen una correlaci√≥n mayor al umbral
upper = corr_matrix.where(np.triu(np.ones(corr_matrix.shape), k=1).astype(bool))

# Hacemos la ista para mantener las columnas a eliminar
cols_to_drop = set()

# Iterar a trav√©s de la matriz de correlaci√≥n para identificar variables altamente correlacionadas
for column in upper.columns:
    if any(upper[column] > threshold):
        cols_to_drop.add(column)

print(f"Columnas a eliminar por alta correlaci√≥n: {cols_to_drop}")
```

```python
# Eliminamos las columnas altamente correlacionadas
data_total_cleaned = data_total_cleaned.drop(columns=cols_to_drop)
```

```python
data_total_cleaned.sample(5)
```
